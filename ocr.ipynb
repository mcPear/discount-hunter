```python
from PIL import Image
import pytesseract
import argparse
import cv2
import os
```

```python
def ocr(filename):
    image = cv2.imread(filename)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    gray = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1] #todo try to separate white and black text using fixed threshold
    bins = [gray, 255-gray]
    bins = [cv2.medianBlur(img, 3) for img in bins]
    texts = [pytesseract.image_to_string(Image.fromarray(img)) for img in bins]
    return texts[0]+texts[1]
```

```python
dir_path = "kaufland"
```

```python
from os import listdir
from os.path import isfile, join
for f in listdir(dir_path):
    filepath = join(dir_path, f)
    text = ocr(filepath)
    print(f,"\n-------------------\n", text)
```
